```python
## using inline plots
%matplotlib inline

## importing the necessary modules
import numpy as np
import matplotlib.pyplot as plt
import pylab as py
import ap22_module as main
import uncertainties as unc
import uncertainties.unumpy as unp
from scipy.optimize import curve_fit
import matplotlib as mpl; mpl.rc('text', usetex=True) ## if latex complains about anything, re-run the cell and hope for the best ;)
```

1a) $\mu A741$ als Gleichspannungsverstärker
===

```python
## loading the first data table
U1, Ua = main.load_data(".data/tab1.txt")

## errors: Ua +/- 2 Digits
Ua_err = 0.02
Ua = unp.uarray(Ua, Ua_err)

## fitting section
def gain(x, v, c):
    return v*x+c

## passing U1 in volts not milivolts for correct value of v, c
popt, pcov = curve_fit(gain, U1*10**(-3), unp.nominal_values(Ua), sigma=unp.std_devs(Ua))

## errors of the optimal parameters:
popt_err = np.sqrt(np.diag(pcov))

## setting up the variables
v, c = popt

## setting up a string for the values
val_str = '$v=' + str(v)[:7] + '\pm' + str(popt_err[0])[:5] + '\n' + '$c=' + str(c)[:7] + '\pm' + str(popt_err[1])[:5] + '\ \mathrm{[V]}$'

## plotting section
fig = plt.figure(figsize=(12,9))
fig.subplots_adjust(hspace=.3)
ax = fig.add_subplot(211)
ax.errorbar(U1, unp.nominal_values(Ua), yerr=unp.std_devs(Ua), fmt='r.', label='Messwerte')
ax.plot(U1, gain(U1*10**(-3), v, c), 'k--', label='Linearer Fit: $U_A = v*U_1+c$')
ax.set_xlabel('$U_1\ \mathrm{[mV]}$')
ax.set_ylabel('$U_A\ \mathrm{[V]}$')
ax.set_title('$Plot\ 1: Gleichspannungsverst \\"{a}rkung\ U_A = f(U_1)$')
ax.grid(axis='y')
ax.text(125, 1.2, val_str)
ax.legend()

######## 2nd data set
## loading the second data table
U1, Ua = main.load_data(".data/tab2.txt")

## errors: Ua +/- 2 Digits
Ua_err = 0.02
Ua = unp.uarray(Ua, Ua_err)

## passing U1 in volts not milivolts for correct value of v, c
## not using the first two and last two values, because the op-amp is in saturation:
popt, pcov = curve_fit(gain, (U1[2:-2])*10**(-3), unp.nominal_values(Ua[2:-2]), sigma=unp.std_devs(Ua[2:-2]))

## errors of the optimal parameters:
popt_err = np.sqrt(np.diag(pcov))

## setting up the variables
v, c = popt

## setting up a string for the values
val_str = '$v=' + str(v)[:7] + '\pm' + str(popt_err[0])[:5] + '\n' + '$c=' + str(c)[:7] + '\pm' + str(popt_err[1])[:5] + '\ \mathrm{[V]}$'

## plotting section
ax = fig.add_subplot(212)
ax.errorbar(U1, unp.nominal_values(Ua), yerr=unp.std_devs(Ua), fmt='r.', label='Messwerte')
ax.plot(U1, gain(U1*10**(-3), v, c), 'k--', label='Linearer Fit: $U_A = v*U_1+c$')
ax.set_xlabel('$U_1\ \mathrm{[mV]}$')
ax.set_ylabel('$U_A\ \mathrm{[V]}$')
ax.set_title('$Plot\ 2: Gleichspannungsverst \\"{a}rkung\ U_A = f(U_1)$')
ax.grid(axis='y')
ax.text(125, 12, val_str)
ax.legend()

##save the figure
#fig.savefig('.data/plot1_2.pdf', format='pdf', dpi=200)
```

1b) $\mu A741$ als Wechselspannungsverstärker

```python
## loading the first data table
Ug, Ua = main.load_data(".data/tab3.txt")
U1 = Ug/10.

## errors: Ua +/- 2 Digits
Ua_err = 0.02
Ua = unp.uarray(Ua, Ua_err)

## passing Ug in milivolts for correct value of v, c
popt, pcov = curve_fit(gain, U1, unp.nominal_values(Ua), sigma=unp.std_devs(Ua))

## errors of the optimal parameters:
popt_err = np.sqrt(np.diag(pcov))

## setting up the variables
v, c = popt

## setting up a string for the values
val_str = '$v=' + str(v)[:5] + '\pm' + str(popt_err[0])[:5] + '$\n$' + 'c=' + str(c)[:5] + '\pm' + str(popt_err[1])[:5] + '\ \mathrm{[V]}$'

## plotting section
fig = plt.figure(figsize=(12,9))
fig.subplots_adjust(hspace=.3)
ax = fig.add_subplot(211)
ax.errorbar(U1, unp.nominal_values(Ua), yerr=unp.std_devs(Ua), fmt='r.', label='Messwerte')
ax.plot(U1, gain(U1, v, c), 'k--', label='Linearer Fit: $U_A = v*U_1+c$')
ax.set_xlabel('$U_1\ \mathrm{[mV_{PP}]} = \\frac{U_G}{10} $')
ax.set_ylabel('$U_A\ \mathrm{[mV_{PP}]}$')
ax.set_title('$Plot\ 3: Wechselspannungsverst \\"{a}rkung\ U_A = f(U_1)$')
ax.grid(axis='y')
ax.set_xlim(10, 85)
ax.text(11, 51, val_str)
ax.legend(loc=0)

############# 2nd data set
## loading the second data table
Ug, Ua = main.load_data(".data/tab4.txt")
U1 = Ug/10.

## errors: Ua +/- 2 Digits
Ua_err = 0.02
Ua = unp.uarray(Ua, Ua_err)

## passing Ug in milivolts for correct value of v, c
popt, pcov = curve_fit(gain, U1, unp.nominal_values(Ua), sigma=unp.std_devs(Ua))

## errors of the optimal parameters:
popt_err = np.sqrt(np.diag(pcov))

## setting up the variables
v, c = popt

## setting up a string for the values
val_str = '$v=' + str(v)[:5] + '\pm' + str(popt_err[0])[:5] + '$\n$' + 'c=' + str(c)[:5] + '\pm' + str(popt_err[1])[:5] + '\ \mathrm{[V]}$'

## plotting section
ax = fig.add_subplot(212)
ax.errorbar(U1, unp.nominal_values(Ua), yerr=unp.std_devs(Ua), fmt='r.', label='Messwerte')
ax.plot(U1, gain(U1, v, c), 'k--', label='Linearer Fit: $U_A = v*U_1+c$')
ax.set_xlabel('$U_1\ \mathrm{[mV_{PP}]} = \\frac{U_G}{10}$')
ax.set_ylabel('$U_A\ \mathrm{[mV_{PP}]}$')
ax.set_title('$Plot\ 4: Wechselspannungsverst \\"{a}rkung\ U_A = f(U_1)$')
ax.grid(axis='y')
ax.set_xlim(10, 85)
ax.text(11, 31, val_str)
ax.legend(loc=0)

## save the figure
#fig.savefig('.data/plot3_4.pdf', format='pdf', dpi=200)
```

2) Freqenzgänge bei verschiedenen Gegenkopplungen
===

```python
## loading the data tables, f stays the same
f, Ua1 = main.load_data(".data/tab5.txt")
f, Ua2 = main.load_data(".data/tab6.txt")
f, Ua3 = main.load_data(".data/tab7.txt")
f, Ua4 = main.load_data(".data/tab8.txt")
f, Ua5 = main.load_data(".data/tab9.txt")

## errors: Ua +/- 2 Digits
Ua_err = 0.02
Ua1 = unp.uarray(Ua1, Ua_err)
Ua2 = unp.uarray(Ua2, Ua_err)
Ua3 = unp.uarray(Ua3, Ua_err)
Ua4 = unp.uarray(Ua4, Ua_err)
Ua5 = unp.uarray(Ua5, Ua_err)

## plotting section
fig = plt.figure(figsize=(12,9))
ax = fig.add_subplot(111)
ax.errorbar(f, unp.nominal_values(Ua1), yerr=unp.std_devs(Ua1), fmt='r.', label='$R_G=680\mathrm{k}\Omega,\ U_G=0.3\mathrm{V_{PP}}$')
ax.errorbar(f, unp.nominal_values(Ua2), yerr=unp.std_devs(Ua2), fmt='b.', label='$R_G=274\mathrm{k}\Omega,\ U_G=0.3\mathrm{V_{PP}}$')
ax.errorbar(f, unp.nominal_values(Ua3), yerr=unp.std_devs(Ua3), fmt='g.', label='$R_G=48.7\mathrm{k}\Omega,\ U_G=0.96\mathrm{V_{PP}}$')
ax.errorbar(f, unp.nominal_values(Ua4), yerr=unp.std_devs(Ua4), fmt='y.', label='$Z_G=(48.7\mathrm{k}\Omega||560\mathrm{pF}),\ U_G=1\mathrm{V_{PP}}$')
ax.errorbar(f, unp.nominal_values(Ua5), yerr=unp.std_devs(Ua5), fmt='k.', label='$R_G=48.7\mathrm{k}\Omega,\ C_{Eingang}=47\mathrm{nF},\ U_G=1\mathrm{V_{PP}}$')
ax.set_xlabel('$\mathrm{log}(f)\ \mathrm{[Hz]}$')
ax.set_ylabel('$\mathrm{log}(U_A)\ \mathrm{[mV_{PP}]}$')
ax.set_title('$Plot\ 5: Freqenzgang\ verschiedener\ Gegenkopplungen$')
ax.grid(axis='both')
ax.set_xscale('log')
ax.set_yscale('log')
ax.legend(loc=0)

## save the figure
#fig.savefig('.data/plot5.pdf', format='pdf', dpi=200)
```
