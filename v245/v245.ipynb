```python
## using inline plots
%matplotlib inline

## importing the necessary modules
import numpy as np
import matplotlib.pyplot as plt
import pylab as py
import ap22_module as main
import uncertainties as unc
import uncertainties.unumpy as unp
from scipy.optimize import curve_fit
import matplotlib as mpl; mpl.rc('text', usetex=True) ## if latex complains about anything, re-run the cell and hope for the best ;)
```

1) Induktionsgesetz
===

```python
## loading the data for Plot1: Um(f)
f, Um = main.load_data('.data/tab1.dat')
df = np.array([0.3, 0.3, 0.1, 0.1, 0.1])
dUm = 0.05

## assigning the values with the errors
f = unp.uarray(f, df)
Um = unp.uarray(Um, dUm)

## linear fit plot 1
def lin_func(x,k,d):
    return k*x+d

popt, pcov = curve_fit(lin_func, unp.nominal_values(f), unp.nominal_values(Um), sigma=unp.std_devs(Um))
popt_err = np.sqrt(np.diag(pcov))
print('k,d & err: plot1:',popt, popt_err)

## plotting section plot1
fig = plt.figure(figsize=(12,9))
ax = fig.add_subplot(211)
ax.errorbar(unp.nominal_values(f), unp.nominal_values(Um), yerr=unp.std_devs(Um), xerr=unp.std_devs(f), fmt='r.', label='Messwerte')
ax.plot(unp.nominal_values(f), lin_func(unp.nominal_values(f), *popt), 'k--', label='Linearer Fit: $U_m=f*k+c$')
ax.set_xlabel('$Frequenz\ f\ \mathrm{[Hz]}$')
ax.set_ylabel('$Induktionsspannung\ U_m\ \mathrm{[V]}$')
ax.set_title('$Plot\ 1:\ Induktionsspannung\ vs.\ Frequenz$')
ax.grid(axis='both')
ax.text(2.1, 4.1, '$k=0.3558\pm0.0048 \mathrm{[\\frac{V}{Hz}]}$\n$d=-0.0973\pm0.0477\mathrm{[V]}$')
ax.legend(loc=0)

########################################################################################################

## loading the data for Plot2: Um(f)
I, Um = main.load_data('.data/tab2.dat')
dI = 0.1
dUm = 0.05

## assigning the values with the errors
I = unp.uarray(I, dI)
Um = unp.uarray(Um, dUm)

popt, pcov = curve_fit(lin_func, unp.nominal_values(I), unp.nominal_values(Um), sigma=unp.std_devs(Um))
popt_err = np.sqrt(np.diag(pcov))
print('k,d & err: plot2:',popt, popt_err)

## plotting section plot2
#fig = plt.figure(figsize=(12,9))
ax = fig.add_subplot(212)
ax.errorbar(unp.nominal_values(I), unp.nominal_values(Um), yerr=unp.std_devs(Um), xerr=unp.std_devs(I), fmt='r.', label='Messwerte')
ax.plot(unp.nominal_values(I), lin_func(unp.nominal_values(I), *popt), 'k--', label='Linearer Fit: $U_m=I*k+c$')
ax.set_xlabel('$Strom\ I\ \mathrm{[A]}$')
ax.set_ylabel('$Induktionsspannung\ U_m\ \mathrm{[V]}$')
ax.set_title('$Plot\ 2:\ Induktionsspannung\ vs.\ Strom$')
ax.grid(axis='both')
ax.text(0.1, 2.83, '$k=0.8116\pm0.0086 \mathrm{[\\frac{V}{A}]}$\n$d=0.2083\pm0.0246\mathrm{[A]}$')
ax.legend(loc=0)
```
